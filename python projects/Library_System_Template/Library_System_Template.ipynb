```python
class Library:
    def __init__(self , books_list , name_of_lirary):
        self.books_list = books_list
        self.name_of_lirary = name_of_lirary
        self.lended_books = []
    
    def available_books(self):
        print (f"available books can you lend : {self.books_list}")
    
    def books_lending(self , book_name):
        if book_name.title() in self.lended_books:
            print("sorry you can't lend that now")
        elif book_name.title() not in self.books_list:
            print("sorry you can't lend this book not available")
        else :
            self.lended_books.append(book_name.title())
            print(f"Ok you lended {book_name.title()} now")
            
    def add(self , add_book):
        if add_book.title() in self.books_list:
            print("can't add this book available")
        else :
            print(f"thank's for adding {add_book}")
            self.books_list.append(add_book.title())
            
    def borrowed_book(self , borrow_book):
        if borrow_book.title() in self.lended_books:
            self.lended_books.remove(borrow_book.title())
        print(f"thank's you borrow {borrow_book.title()} book")
```

```python
Harry = Library(['Python', 'Rich Daddy Poor Daddy', 'Harry Potter', 'C++ Basics', 'Algorithms by CLRS'], "CodeWithHarry")

print(f"Welcome to {Harry.name_of_lirary} Library!")

while True:
    print("1 : Available Books")
    print("2 : Books Lending")
    print("3 : Add Books")
    print("4 : Borrowed Books")
    print("5 : Exit")
    user = input("Enter the number of your choice: ")

    if user == "1":
        Harry.available_books()
        
    elif user == "2":
        book = input("Enter the name of the book you want to lend: ")
        Harry.books_lending(book)
                           
    elif user == "3":
        book_to_add = input("Enter the name of the book you want to add: ")
        Harry.add(book_to_add)
        
    elif user == "4":
        book_to_return = input("Enter the name of the book you want to return: ")
        Harry.borrowed_book(book_to_return)
        
    elif user == "5":
        print(f"Thank you, see you soon at {Harry.name_of_lirary}!")
        break
    else:
        print("Invalid choice. Please try again.")
```
